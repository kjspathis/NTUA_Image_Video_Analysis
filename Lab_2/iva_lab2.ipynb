## Θεωρητικό Μέρος

### 1)

Η **Rectified Linear Unit (ReLU)** είναι μια μη γραμμική συνάρτηση ενεργοποίησης ευρέως διαδεδομένη στη μηχανική μάθηση, η οποία ορίζεται σύμφωνα με τη σχέση : 

<h3 align="center">f(x) = max(0,x)</h3>

Η λειτουργία της ReLU συνοψίζεται στην ακόλουθη λογική. 

_Αν κάποιος νευρώνας παράγει σαν αποτέλεσμα θετική τιμή, τότε θα πραγματοποιηθεί εκπαίδευση στο συγκεκριμένο νευρώνα, διαφορετικά παραλείπεται._ 

Η ReLU εφαρμόζεται στα κρυφά επίπεδα των βαθιών συνελικτικών νευρωνικών δικτύων, ξεχωριστά σε κάθε pixel του feautures map, που παράχθηκε από το στάδιο της συνέλιξης. Η ευρεία χρήση της ReLU και η επιλογή της σε σχέση με άλλες, όπως πχ η tanh, οφείλεται στη δυνατότητά της να εκπαιδευτεί ταχύτερα με τεχνικές gradient descent, αυξάνοντας τις μη γραμμικές ιδιότητες του δικτύου χωρίς να επηρεάζει τα recieptive fields του. Επιπλέον, πετυχαίνει αραιή ενεργοποίηση νευρώνων, μιας και ένα μεγάλο ποσοστό των νευρώνων λαμβάνουν μη θετική τιμή, με αποτέλεσμα να παραλείπονται. Με αυτό τον τρόπο μειώνεται η πιθανότητα να εμφανιστούν φαινόμενα όπως vanishing gradient και overfitting.  

### 2)

Σύμφωνα με τα δημοσιευθέντα άρθρα : 
* [Handwritten Digit Recognition with a Back-Propagation Network](https://papers.nips.cc/paper/1989/file/53c3bce66e43be4f209556518c2fcb54-Paper.pdf) των Y. Le Cun, B. Boser, J. S. Denker, D. Henderson, R. E. Howard, W. Hubbard, και L. D. Jackel 
* [ImageNet Classification with Deep Convolutional Neural Networks](https://proceedings.neurips.cc/paper/2012/file/c399862d3b9d6b76c8436e924a68c45b-Paper.pdf) των Alex Krizhevsky, Ilya Sutskever και Geoffrey E. Hinton
* [Very Deep Convolutional Networks for Large-Scale Image Recognition](https://arxiv.org/pdf/1409.1556.pdf) των Karen Simonyan και Andrew Zisserman

Τα τρία νευρωνικά δίκτυα, που παρουσιάζονται στα παραπάνω άρθρα, είναι το LeNet, το AlexNet και το VGG. Συγκεκριμένα, 

|                             | LeNet       | AlexNet       |  VGG     |
| :---                        |    :----:   |         :---: | :---:    |
| **Επίπεδα**                 | 7 | 8 | 9 |
| **Μέγεθος Φίλτρων (kernel-sized)** | 5x5x3 | 3x3x3 & 5x5x3 & 11x11x3 | 3×3x3 |
| **Συνάρτηση Ενεργοποίησης** | Sigmoid | ReLU | ReLU |
| **Πλήθος Παραμέτρων**       | 60.000 | 61.000.000| 138.000.000 |
| **Τεχνικές Pooling**        | average-pooling of size 2x2 with a stride of 2 | max-pooling of size 3×3 with a stride of 2 | max-pooling of size 2 × 2 with strides of 2 |
| **Τεχνικές Dropout**        | - | rate 0.5 | - |


Από τα παραπάνω φαίνεται ότι :
* Το **LeNet** έχει τα λιγότερα επίπεδα και τις λιγότερες παραμέτρους. Παράλληλα, σαν συνάρτηση ενεργοποίησης χρησιμοποιεί τη σιγμοειδή και σαν τεχνική pooling χρησιμοποιεί average, σε αντίθεση με τα άλλα μοντέλα που χρησιμοποιούν τη ReLU και max pooling.
* Το **AlexNet** χρησιμοποιεί τριών διαστάσεων φίλτρα διαφορετικά ανα επίπεδο, μιας λαμβάνει ως είσοδο εικόνες μεγάλων διαστάσεων (224x224 pixels), που χρειάζονται επιπρόσθετη επεξεργασία. Επιπλέον, μόνο το AlexNet χρησιμοποιεί Dropout για την αποφυγή του overfitting. 
* Το **VGG** είναι το μεγαλύτερο σε βάθος δίκτυο από τα παραπάνω δίκτυα, καθώς διαθέτει τα περισσότερα επίπεδα και τις περισσότερες παραμέτρους, με αποτέλεσμα να χρειάζεται και την περισσότερη ώρα για την εκπαίδευσή του.

Τα χαρακτηριστικά αυτών των τριών δικτύων, που συνέβαλαν στη σταδιακή αύξηση της ακρίβειας ταξινόμησης είναι :
* Η **χρήση συναρτήσεων ενεργοποίησης**, με τη βοήθεια των οποίων μοντελοποιείται η μη γραμμικότητα του προβλήματος
* Η **πολυεπίπεδη αρχιτεκτονική** των δικτύων, κάνοντας εφικτό να εισαχθούν στα μοντέλα απευθείας εικόνες και να γίνει σε αυτά η εξαγωγή χαρακτηριστικών (κάτι που πριν την αρχιτεκτονική των CNN γινόταν στο στάδιο του Data Processing) και ύστερα η ταξινόμησή τους σε κλάσεις
* Η **χρήση τεχνικών Pooling και Dropout**, με τη βοήθεια των οποίων περιορίστηκε η εμφάνιση φαινομένων όπως overfitting και vanishing gradient.

```
# helper functions

# select from from_list elements with index in index_list
def select_from_list(from_list, index_list):
  filtered_list= [from_list[i] for i in index_list]
  return(filtered_list)

# append in filtered_list the index of each element of unfilterd_list if it exists in in target_list
def get_ds_index(unfiliterd_list, target_list):
  index = 0
  filtered_list=[]
  for i_ in unfiliterd_list:
    if i_[0] in target_list:
      filtered_list.append(index)
    index += 1
  return(filtered_list)
```

#### Class mapping 
Tα labels, που διαθέτουμε αντιστοιχούν μόνο σε 20 κατηγορίες, χωρίς όμως αυτό να συνεπάγεται ότι τα indexes είναι συνεχόμενα. Για αυτό το λόγο γίνεται η ακόλουθη αντιστοίχιση:

```
# Map Image Labels to [0,19]
def Dict_Creator(Labels):
    Dict = {}
    c=0
    for Label in Labels:
        if Label not in Dict.keys():
          Dict[Label] = c
          c+=1
    return Dict

def Map_Creator(Labels, Dict): 
    return [Dict[x] for x in Labels]

print("The labels before:",y_train[:10])

# Create Label Dict
label_dict = Dict_Creator(y_train.reshape(1,-1)[0])
# Apply mapping
y_train = Map_Creator(y_train.reshape(1,-1)[0], label_dict)
y_test = Map_Creator(y_test.reshape(1,-1)[0], label_dict)
y_val = Map_Creator(y_val.reshape(1,-1)[0], label_dict)

y_train = np.array([[i] for i in y_train])
y_test = np.array([[i] for i in y_test])
y_val = np.array([[i] for i in y_val])

print("The labels after:",y_train[:10])
```

#### Διαχείριση μνήμης (TFRecord)
Η φόρτωση δεδομένων με τον τρόπο που το κάναμε παραπάνω στο απλό παράδειγμα υλοποίησης είναι πολύ βολική αλλά δεν είναι αποτελεσματική ως προς τη διαχείριση της μνήμης. Συγκεκριμένα, με τον τρόπο αυτό, τα δεδομένα αποθηκεύονται απευθείας σε μεταβλητές, οι οποίες όλες μαζί καταλαμβάνουν τη RAM της CPU ή της GPU, κάτι που κάνει αδύνατη τη διαχείριση μεγάλων datasets ή τον μεταχηματισμό των δεδομένων όπως όταν κάνουμε αύξηση δεδομένων (data augmentation).

Για να παρακαμφθεί αυτό το πρόβλημα, υπάρχει η δυνατότητα της σειριοποίησης των δεδομένων (serialization) και της αποθήκευσής τους σε αρχεία μεσαίου μεγέθους (κάποιων MB) τα οποία μπορούν να αναγνωστούν γραμμικά.

Το φορμάτ TFRecord είναι ένα φορμάτ που επιτρέπει την αποθήκευση σειράς δυαδικών εγγραφών. Διαβάστε σχετικά για το [TFRecord and tf.Example](https://www.tensorflow.org/tutorials/load_data/tfrecord) και [tf.data: Build TensorFlow input pipelines](https://www.tensorflow.org/guide/data). 

Σημειώστε ότι με τη μέθοδο αυτή θα πρέπει να γίνει import η `tensorflow_datasets` και να χρησιμοποιήσουμε την `tfds.load` ώστε να αποθηκευθεί το σύνολο δεδομένων σε αρχεία tfrecord στο δίσκο (δείτε [εδώ](https://colab.research.google.com/github/tensorflow/datasets/blob/master/docs/overview.ipynb) ένα παράδειγμα). Φυσικά μπορούμε να μετατρέψουμε και τα πρωτογενή δεδομένα (raw data) του dataset όπως αρχεία jpg σε φορματ tfrecord όπως [εδώ](https://towardsdatascience.com/working-with-tfrecords-and-tf-train-example-36d111b3ff4d).


```
# according to 
# ## https://towardsdatascience.com/a-practical-guide-to-tfrecords-584536bc786c ##

def _bytes_feature(value):
  """Returns a bytes_list from a string / byte."""
  if isinstance(value, type(tf.constant(0))): # if value ist tensor
    value = value.numpy() # get value of tensor
  
  return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
  """Returns a floast_list from a float / double."""
  return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
  """Returns an int64_list from a bool / enum / int / uint."""
  return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))

def serialize_array(array):
  array = tf.io.serialize_tensor(array)
  return array

def parse_single_image(image, label):
  
  #define the dictionary -- the structure -- of our single example
  data = {
        'height' : _int64_feature(image.shape[0]),
        'width' : _int64_feature(image.shape[1]),
        'depth' : _int64_feature(image.shape[2]),
        'raw_image' : _bytes_feature(serialize_array(image)),
        'label' : _int64_feature(label)
    }
  #create an Example, wrapping the single features
  out = tf.train.Example(features=tf.train.Features(feature=data))

  return out

def write_images_to_tfr_short(images, labels, filename:str="images"):
  filename= filename+".tfrecords"
  writer = tf.io.TFRecordWriter(filename) #create a writer that'll store our data to disk
  count = 0

  for index in range(len(images)):

    #get the data we want to write
    current_image = images[index] 
    current_label = labels[index]

    out = parse_single_image(image=current_image, label=current_label)
    writer.write(out.SerializeToString())
    count += 1

  writer.close()
  print(f"Wrote {count} elements to TFRecord")
  return count

def parse_tfr_element(element):
  #use the same structure as above; it's kinda an outline of the structure we now want to create
  data = {
      'height': tf.io.FixedLenFeature([], tf.int64),
      'width':tf.io.FixedLenFeature([], tf.int64),
      'label':tf.io.FixedLenFeature([], tf.int64),
      'raw_image' : tf.io.FixedLenFeature([], tf.string),
      'depth':tf.io.FixedLenFeature([], tf.int64),
    }

    
  content = tf.io.parse_single_example(element, data)
  
  height = content['height']
  width = content['width']
  depth = content['depth']
  label = content['label']
  raw_image = content['raw_image']
  
  
  #get our 'feature'-- our image -- and reshape it appropriately
  feature = tf.io.parse_tensor(raw_image, out_type=tf.int16)
  feature = tf.reshape(feature, shape=[height,width,depth])
  return (feature, label)

def get_dataset_small(filename):
  #create the dataset
  dataset = tf.data.TFRecordDataset(filename)

  #pass every single feature through our mapping function
  dataset = dataset.map(
      parse_tfr_element
  )
    
  return dataset
```

##### **LeNet Model** 
Αυτό το νευρωνικό δίκτυο αποτελείται από τα ακόλουθα επιπέδα :  
* 2 Convolutional Layers με μέγεθος πυρήνα 5x5 και πλήθος φίλτρων 6 και 16 αντιστοίχως

##### **AlexNet**
Αυτό το νευρωνικό δίκτυο αποτελείται από τα ακόλουθα επιπέδα :  
* 5 Convolutional layers με μέγεθος πυρήνα 11x11, 5x5, 3x3, 3x3, 3x3 και πλήθος φίλτρων 96, 256, 384, 384, 256 αντιστοίχως
* 3 MaxPooling layers με μέγεθος πυρήνα 3x3 και stride 2
* 2 Dropout layers με 0.5 dropout rate
* 3 Fully Conected layers με 4096, 4096 και 20 νευρώνες

```
#####################
""" AlexNet Class """
#####################

import tensorflow as tf
#from d2l import tensorflow as d2l
sparse = tf.keras.losses.SparseCategoricalCrossentropy()
logcosh = tf.keras.losses.LogCosh()
#alexnet_in = (224,224,3)
alexnet_in = (32,32,3)

# Following the basic implementation of a AlexNet model according to this site
# https://d2l.ai/chapter_convolutional-modern/alexnet.html
# i made a class to create, train and evaluate AlexNet model with different parameters

""" NOTES about AlexNet
        # Here, we use a larger 11 x 11 window to capture objects. At the same
        # time, we use a stride of 4 to greatly reduce the height and width of
        # the output. Here, the number of output channels is much larger than
        # that in LeNet

        # Make the convolution window smaller, set padding to 2 for consistent
        # height and width across the input and output, and increase the
        # number of output channels

        # Use three successive convolutional layers and a smaller convolution
        # window. Except for the final convolutional layer, the number of
        # output channels is further increased. Pooling layers are not used to
        # reduce the height and width of input after the first two
        # convolutional layers

        # Here, the number of outputs of the fully-connected layer is several
        # times larger than that in LeNet. Use the dropout layer to mitigate
        # overfitting
  """

class AlexNet_Models_Creator:
  def __init__(self, input_shape, output_shape):
    self.input_shape = input_shape # image shape (224,224,3)
    self.output_shape = output_shape # number of classes 20
    self.AlexNet_model = None
    self.history = None
  
  def create_model(self, activation_type, pooling_size, kernel_size, filters, drop_choser, stride):
    ## 1st step : upscale images, since AlexNet has an input of 224x224
    self.AlexNet_model = tf.keras.models.Sequential([tf.keras.layers.experimental.preprocessing.Resizing(224, 224, interpolation="bilinear", input_shape=self.input_shape),
                                                     tf.keras.layers.Conv2D(filters=filters[0], kernel_size=kernel_size[0], strides=stride[0], activation=activation_type),
                                                     tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                     tf.keras.layers.Conv2D(filters=filters[1], kernel_size=kernel_size[1], padding='same', activation=activation_type),
                                                     tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                     tf.keras.layers.Conv2D(filters=filters[2], kernel_size=kernel_size[2], padding='same', activation=activation_type),
                                                     tf.keras.layers.Conv2D(filters=filters[3], kernel_size=kernel_size[3], padding='same', activation=activation_type),
                                                     tf.keras.layers.Conv2D(filters=filters[4], kernel_size=kernel_size[4], padding='same', activation=activation_type),
                                                     tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                     tf.keras.layers.Flatten(),
                                                     tf.keras.layers.Dense(4096, activation=activation_type),
                                                     tf.keras.layers.Dropout(drop_choser),
                                                     tf.keras.layers.Dense(4096, activation=activation_type),
                                                     tf.keras.layers.Dropout(drop_choser),
                                                     tf.keras.layers.Dense(self.output_shape)])

    return self.AlexNet_model 

  def train_model(self, train_images, train_labels, val_images, val_labels, num_epochs, batch_size,  optimizer_choser, loss_choser):
    if(loss_choser=='sparse'):
      self.AlexNet_model.compile(optimizer=optimizer_choser, loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
      self.history = self.AlexNet_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    if(loss_choser=='logcosh'):
      self.AlexNet_model.compile(optimizer=optimizer_choser, loss=tf.keras.losses.LogCosh(reduction="auto", name="log_cosh"), metrics=['accuracy'])
      self.history = self.AlexNet_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    if(loss_choser=='KLdiv'):
      self.AlexNet_model.compile(optimizer=optimizer_choser, loss=tf.keras.losses.KLDivergence(), metrics=['accuracy'])
      self.history = self.AlexNet_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    return self.AlexNet_model, self.history 
  
  def evaluate_model(self, test_images, test_labels):
    
    doublePlotarisma(self.history.history['accuracy'], self.history.history['val_accuracy'], 
                     'accuracy', 'val_accuracy', 'AlexNet model', 'Epoch', 'Accuracy')

    test_loss, test_acc = self.AlexNet_model.evaluate(test_images,  test_labels, verbose=2)
    print("The loss of this AlexNet model is: "+str(test_loss))
    print("The accuracy of this AlexNet model is: "+str(test_acc))
    
    return test_loss, test_acc
```

##### ***VGG***

Αυτό το νευρωνικό δίκτυο αποτελείται από δύο μέρη:
* το πρώτο δομείται από Convolutional και Pooling layers 
* το δεύτερο δομείται από Fully Connected layeers

Το VGG κατασκευάζεται διαδοχικά από blocks, καθένα από τα οποία περιλαμβάνει μια ακολουθία convolutional layers και ένα maxpooling layer.

```
#################
""" VGG Class """
#################

import tensorflow as tf
#from d2l import tensorflow as d2l
sparse = tf.keras.losses.SparseCategoricalCrossentropy()
logcosh = tf.keras.losses.LogCosh()
#vgg_in = (224,224,3)
vgg_in = (32,32,3)
# Following the basic implementation of a VGG model according to this site
# https://d2l.ai/chapter_convolutional-modern/vgg.html
# i made a class to create, train and evaluate VGG model with different parameters

vgg_conv_arch = ((1, 64), (1, 128), (2, 256), (2, 512), (2, 512))

class VGG_Models_Creator:
  """ NOTES about VGG
        
  """

  def __init__(self, input_shape, output_shape):
    self.input_shape = input_shape # image shape (224,224,3)
    self.output_shape = output_shape # number of classes 20
    self.VGG_model = None
    self.history = None

  def vgg_block(self, num_convs, num_channels, activation_type, pooling_size, kernel_size, stride):
    
    blk = tf.keras.models.Sequential()
    
    for _ in range(num_convs):
      blk.add(tf.keras.layers.Conv2D(num_channels, kernel_size=kernel_size, padding='same', activation=activation_type))
    
    blk.add(tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride))
    
    return blk
  
  def vgg(self, conv_arch, activation_type, pooling_size, kernel_size, drop_choser, stride):
    net = tf.keras.models.Sequential()
    net.add(tf.keras.layers.experimental.preprocessing.Resizing(224, 224, interpolation="bilinear", input_shape=self.input_shape))
    
    # The convulational part
    for (num_convs, num_channels) in conv_arch:
      net.add(self.vgg_block(num_convs, num_channels, activation_type, pooling_size, kernel_size, stride))
    
    # The fully-connected part
    net.add(tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                        tf.keras.layers.Dense(4096, activation='relu'),
                                        tf.keras.layers.Dropout(drop_choser),
                                        tf.keras.layers.Dense(4096, activation='relu'),
                                        tf.keras.layers.Dropout(drop_choser),
                                        tf.keras.layers.Dense(self.output_shape, activation='softmax')]))
    return net
  
  def create_model(self, conv_arch, activation_type, pooling_size, kernel_size, drop_choser, stride):
    
    self.VGG_model = self.vgg(conv_arch, activation_type, pooling_size, kernel_size, drop_choser, stride)

    return self.VGG_model 

  def train_model(self, train_images, train_labels, val_images, val_labels, num_epochs, batch_size,  optimizer_choser, loss_choser):
    if(loss_choser=='sparse'):
      self.VGG_model.compile(optimizer=optimizer_choser, loss= tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
      self.history = self.VGG_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    if(loss_choser=='hinge'):
      self.VGG_model.compile(optimizer=optimizer_choser, loss=tf.keras.losses.Hinge(), metrics=['accuracy'])
      self.history = self.VGG_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    if(loss_choser=='logcosh'):
      self.VGG_model.compile(optimizer=optimizer_choser, loss=tf.keras.losses.LogCosh(reduction="auto", name="log_cosh"), metrics=['accuracy'])
      self.history = self.VGG_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    return self.VGG_model, self.history
  
  def evaluate_model(self, test_images, test_labels):
    
    doublePlotarisma(self.history.history['accuracy'], self.history.history['val_accuracy'], 
                     'accuracy', 'val_accuracy', 'VGG model', 'Epoch', 'Accuracy')

    test_loss, test_acc = self.VGG_model.evaluate(test_images,  test_labels, verbose=2)
    print("The loss of this VGG model is: "+str(test_loss))
    print("The accuracy of this VGG model is: "+str(test_acc))
    
    return test_loss, test_acc
```

##### **MyCNN**
Αυτό το νευρωνικό δίκτυο είναι μια προσωπική υλοποίηση βασισμένη στον επιλεκτικό συνδυασμό των καλύτερων αρχιτεκτονικών των παραπάνω μοντέλων (LeNet, AlexNet, VGG), όπως παρατηρήθηκαν από τις επιδόσεις τους. Συγκεκριμένα, το μοντέλο αυτό αποτελείται από τα ακόλουθα επιπέδα :  
* 1 layer για αλλαγή των διαστάσεων των εικόνων, καθώς χρησιμοποιήθηκαν μεγάλων διαστάσεων πυρήνες
* 5 Convolutional layers με μέγεθος πυρήνα 11x11, 5x5, 3x3, 3x3, 3x3 και πλήθος φίλτρων 96, 256, 384, 384, 256 αντιστοίχως, κάθε layer ακολουθείται από ένα BatchNormalization layer
* 5 BatchNormalization layers, με momentum ίσο με 0.8, κάθε layer ακολουθείται από ένα MaxPooling layer
* 5 MaxPooling layers με μέγεθος πυρήνα 3x3 και stride 2, κάθε layer ακολουθείται από ένα Convolutional layer
* 5 Fully Conected layers με 4096, 4096 και 20 νευρώνες

```
###################
""" MyCNN Class """
###################

import tensorflow as tf
#from d2l import tensorflow as d2l
sparse = tf.keras.losses.SparseCategoricalCrossentropy()
logcosh = tf.keras.losses.LogCosh()
#mycnn_in = (224,224,3)
mycnn_in = (32,32,3)

class MyCNN_Models_Creator:
  def __init__(self, input_shape, output_shape):
    self.input_shape = input_shape # image shape (224,224,3)
    self.output_shape = output_shape # number of classes 20
    self.MyCNN_model = None
    self.history = None
  
  def create_model(self, activation_type, pooling_size, kernel_size, filters, stride):
    self.MyCNN_model = tf.keras.models.Sequential([tf.keras.layers.experimental.preprocessing.Resizing(224, 224, interpolation="bilinear", input_shape=self.input_shape),
                                                   tf.keras.layers.Conv2D(filters=filters[0], kernel_size=kernel_size[0], strides=stride[0], activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[1], kernel_size=kernel_size[1], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[2], kernel_size=kernel_size[2], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[3], kernel_size=kernel_size[3], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[4], kernel_size=kernel_size[4], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Flatten(),
                                                   tf.keras.layers.Dense(2048, activation=activation_type),
                                                   tf.keras.layers.Dense(1024, activation=activation_type),
                                                   tf.keras.layers.Dense(512, activation=activation_type),
                                                   tf.keras.layers.Dense(256, activation=activation_type),
                                                   tf.keras.layers.Dense(self.output_shape, activation="softmax")])

    return self.MyCNN_model 

  def train_model(self, train_images, train_labels, val_images, val_labels, num_epochs, batch_size,  optimizer_choser, loss_choser):
    
    if(loss_choser=='sparse'):
      self.MyCNN_model.compile(optimizer=optimizer_choser, loss= tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
      self.history = self.MyCNN_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    if(loss_choser=='categorical'):
      self.MyCNN_model.compile(optimizer=optimizer_choser, loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy'])
      self.history = self.MyCNN_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(val_images, val_labels))
    
    return self.MyCNN_model, self.history 
  
  def evaluate_model(self, test_images, test_labels):
    
    doublePlotarisma(self.history.history['accuracy'], self.history.history['val_accuracy'], 
                     'accuracy', 'val_accuracy', 'MyCNN model', 'Epoch', 'Accuracy')

    test_loss, test_acc = self.MyCNN_model.evaluate(test_images,  test_labels, verbose=2)
    print("The loss of this MyCNN model is: "+str(test_loss))
    print("The accuracy of this MyCNN model is: "+str(test_acc))
    
    return test_loss, test_acc
```

##### Πρόωρος τερματισμός (early stopping [tf.keras.callbacks.EarlyStopping](https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStoppinghttps://))

```
#######################################
""" MyCNN Class with Early Stopping """
#######################################


import tensorflow as tf
#from d2l import tensorflow as d2l
sparse = tf.keras.losses.SparseCategoricalCrossentropy()
logcosh = tf.keras.losses.LogCosh()
#mycnn_in = (224,224,3)
mycnn_in = (32,32,3)

EarlyStop = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5)

class MyCNN_Early_Creator:
  def __init__(self, input_shape, output_shape):
    self.input_shape = input_shape # image shape (224,224,3)
    self.output_shape = output_shape # number of classes 20
    self.MyCNN_model = None
    self.history = None
  
  def create_model(self, activation_type, pooling_size, kernel_size, filters, stride):
    self.MyCNN_model = tf.keras.models.Sequential([tf.keras.layers.experimental.preprocessing.Resizing(224, 224, interpolation="bilinear", input_shape=self.input_shape),
                                                   tf.keras.layers.Conv2D(filters=filters[0], kernel_size=kernel_size[0], strides=stride[0], activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[1], kernel_size=kernel_size[1], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[2], kernel_size=kernel_size[2], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[3], kernel_size=kernel_size[3], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Conv2D(filters=filters[4], kernel_size=kernel_size[4], padding='same', activation=activation_type),
                                                   tf.keras.layers.BatchNormalization(momentum=0.8),
                                                   tf.keras.layers.MaxPool2D(pool_size=pooling_size, strides=stride[1]),
                                                   tf.keras.layers.Flatten(),
                                                   tf.keras.layers.Dense(2048, activation=activation_type),
                                                   tf.keras.layers.Dense(1024, activation=activation_type),
                                                   tf.keras.layers.Dense(512, activation=activation_type),
                                                   tf.keras.layers.Dense(256, activation=activation_type),
                                                   tf.keras.layers.Dense(self.output_shape, activation="softmax")])

    return self.MyCNN_model 

  def train_model(self, train_images, train_labels, val_images, val_labels, num_epochs, batch_size,  optimizer_choser, loss_choser):
    
    if(loss_choser=='sparse'):
      self.MyCNN_model.compile(optimizer=optimizer_choser, loss= tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
      self.history = self.MyCNN_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, callbacks = EarlyStop, validation_data=(val_images, val_labels))
    
    if(loss_choser=='categorical'):
      self.MyCNN_model.compile(optimizer=optimizer_choser, loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy'])
      self.history = self.MyCNN_model.fit(train_images, train_labels, epochs=num_epochs, batch_size=batch_size, callbacks = EarlyStop, validation_data=(val_images, val_labels))
    
    return self.MyCNN_model, self.history 
  
  def evaluate_model(self, test_images, test_labels):
    
    doublePlotarisma(self.history.history['accuracy'], self.history.history['val_accuracy'], 
                     'accuracy', 'val_accuracy', 'MyCNN model', 'Epoch', 'Accuracy')

    test_loss, test_acc = self.MyCNN_model.evaluate(test_images,  test_labels, verbose=2)
    print("The loss of this MyCNN model is: "+str(test_loss))
    print("The accuracy of this MyCNN model is: "+str(test_acc))
    
    return test_loss, test_acc
```
